```python
import statsmodels.api as sm
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from sklearn.impute import SimpleImputer



```

```python
df = pd.read_csv("Air_Quality_Monitoring_Data.csv")
df.columns = df.columns.str.lower()

df.head()
```

```python
from datetime import datetime

df['datetime'] = pd.to_datetime(df['datetime'], format='%d/%m/%Y %H:%M')
df['datetime'] = df['datetime'].dt.month.astype(float)
```

```python
df.head()
```

```python
df['time'] = pd.to_datetime(df['time'])
df['time'] = df['time'].dt.hour.astype(float)

df.head()
```

```python
location_mapping = {'Civic': 1, 'Monash': 2, 'Florey': 3}

df['name'] = df['name'].map(location_mapping).astype(float)
df.head()
```

```python
print(df.columns)
```

```python
df.drop(['gps', 'date', 'aqi_co', 'aqi_no2',
        'aqi_o3_1hr', 'aqi_o3_4hr', 'aqi_o3_8hr', 'aqi_pm10', 'aqi_pm2_5'],axis='columns',inplace=True)

df.head()
```

```python
def classify(v):
    if 0 <= v <=33: #Very Good
        return 1
    elif 34 <= v <=66: #Good
        return 2
    elif 67 <= v <=99: #Fair
        return 3
    elif 100 <= v <=149: #Poor
        return 4
    elif 150 <= v <=200: #Very Poor
        return 5
    elif 150 <= v <=200: #Hazardous
        return 6
    
df['aqi_site'] = df['aqi_site'].apply(classify)

df.head()
```

```python
y = df.aqi_site
X = df.drop('aqi_site', axis='columns')
```

```python
# make into method

X.o3_1hr = pd.to_numeric(X.o3_1hr, errors='coerce').astype(float)
X.o3_4hr = pd.to_numeric(X.o3_4hr, errors='coerce').astype(float)
X.o3_8hr = pd.to_numeric(X.o3_8hr, errors='coerce').astype(float)
X.co = pd.to_numeric(X.co, errors='coerce').astype(float)
X.pm10_1_hr = pd.to_numeric(X.pm10_1_hr, errors='coerce').astype(float)
X.pm2_5_1_hr = pd.to_numeric(X.pm2_5_1_hr, errors='coerce').astype(float)
X.pm10 = pd.to_numeric(X.pm10, errors='coerce').astype(float)
X.pm2_5 = pd.to_numeric(X.pm2_5, errors='coerce').astype(float)
X.no2 = pd.to_numeric(X.no2, errors='coerce').astype(float)

X.o3_1hr = X.o3_1hr.fillna(X.o3_1hr.mean())
X.o3_4hr = X.o3_4hr.fillna(X.o3_4hr.mean())
X.o3_8hr = X.o3_8hr.fillna(X.o3_4hr.mean())
X.co = X.co.fillna(X.co.mean())
X.pm10_1_hr = X.pm10_1_hr.fillna(X.pm10_1_hr.mean())
X.pm2_5_1_hr = X.pm2_5_1_hr.fillna(X.pm2_5_1_hr.mean())
X.pm10 = X.pm10.fillna(X.pm10.mean())
X.pm2_5 = X.pm2_5.fillna(X.pm2_5.mean())
X.no2 = X.no2.fillna(X.no2.mean())

X.head()
```

```python
has_nan = y.isna
print(has_nan)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=17)
```

```python
df_train = pd.DataFrame(data=np.column_stack((X_train, y_train)), columns=['name', 'datetime', 'no2', 'o3_1hr', 'o3_4hr', 'o3_8hr', 'co',
       'pm10_1_hr', 'pm2_5_1_hr', 'pm10', 'pm2_5', 'time', 'aqi_site'])
```

```python
model = sm.GLM.from_formula('aqi_site ~ name + datetime + no2 + o3_1hr + o3_4hr + o3_8hr + co + pm10_1_hr + pm2_5_1_hr + pm10 + pm2_5 + time', family=sm.families.Gaussian(), data=df_train)
result = model.fit()
```

```python
df_test = pd.DataFrame(data=np.column_stack((X_test, y_test)), columns=['name', 'datetime', 'no2', 'o3_1hr', 'o3_4hr', 'o3_8hr', 'co',
        'pm10_1_hr', 'pm2_5_1_hr', 'pm10', 'pm2_5', 'time', 'aqi_site'])
y_pred = np.round(result.predict(df_test))
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```
